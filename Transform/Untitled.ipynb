```python
import pandas as pd
import numpy as np
```

```python
data_directory = "../../data_project-it/Data";
carbon_data = "/annual-co2-emissions-per-country.csv";
earning_data = "/filtered_earnings.csv";
```

```python
carbon_df = pd.read_csv(data_directory + carbon_data);
earning_df = pd.read_csv(data_directory + earning_data);
```

```python
## Calculate Correlation Coefficient Between Annual CO2 Emission worldwide and Top 6 Commpany Revenue ##

# Get the unique company symbols to iterate for. #
company_list = list(set(earning_df['symbol']));

# Create a dictionary object to store the data. #
company_earning_avg = {};

# Get the revenue average for each year per company earning. #
# - Start by filtering the dataframe for each company symbol. - #
for i in company_list:
    filtered_earning = earning_df.loc[earning_df['symbol'] == i];
    company_earning_avg[i] = {};
    for idx, col in filtered_earning.iterrows():
        date_year = col['date'][:4];
        if date_year not in company_earning_avg[i].keys():
            company_earning_avg[i][date_year] = [];
        
        company_earning_avg[i][date_year].append(col['revenue']);
        
print(company_earning_avg);
```
